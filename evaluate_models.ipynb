# Evaluaton of YOLO Models

This notebook evaluates to models:
1. Best_yolo_medical.pt
2. yolo11n.pt (Pretrained YOLOv11 model)

We will use the dataset configuration dataset_yolo_final/data.yaml.


```python
!pip install -q ultralytics
import ultralytics
ultralytics.checks()
```

```python
from ultralytics import YOLO

# 1. Load best_yolo_medical.pt
model_best = YOLO('best_yolo_medical.pt')

# 2. Evaluate model_best
print('Evaluating best_yolo_medical.pt...')
metrics_best = model_best.val(data='dataset_yolo_final/data.yaml', imgsz=640, plots=True, project='runs/val', name='best_val')
print(f'mAP50-95 for best_yolo_medical.pt: {metrics_best.box.map:.4f}')
```

```python
# 3. Load YOLOv11 pretrained model (e.g. yolo11n.pt)
model_11 = YOLO('yolo11n.pt')

# 4. Evaluate YOLOv11
print('Evaluating yolo11n.pt...')
metrics_11 = model_11.val(data='dataset_yolo_final/data.yaml', imgsz=640, plots=True, project='runs/val', name='yolo11_val')
print(f'mAP50-95 for yolo11n.pt: {metrics_11.box.map:.4f}')
```
